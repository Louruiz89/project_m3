```python
#Import

import pandas as pd
import numpy as np
```

```python
#Preparo el dataset de training

df_training = pd.read_csv('./../data/diamons_clean.csv')
df_training.head()
```

```python
#Elimino "city"

df_training = df_training.drop(columns='city')
df_training[:2]
```

```python
#Preparo el dataset con nuevos features

df_training['xy'] = df_training['x']/df_training['y']
df_training['xz'] = df_training['x']/df_training['z']
df_training['zy'] = df_training['z']/df_training['y']
df_training['table_depth'] = df_training['table']/df_training['depth']
df_training['carat/dimensons'] = df_training ['carat'] / (df_training['x']* df_training['y'] * df_training['z'])
df_training[:2]
```

```python
#Haciendo el modelo me da un error dónde algunos valores tienden a infinito por lo que elimino esa posibilidad

df_training.replace([np.inf, -np.inf], 0 , inplace=True)
```

```python
#divido el dataset en 2: numérico y categórico

df_num = df_training[["price","carat","depth","table","x","y","z","xy","xz","table_depth","carat/dimensons"]]
df_cat =df_training[["clarity","color","cut"]]
```

```python
#Aplico One Hot enconding en las columnas categóricas

df_cat = ["clarity","color","cut"]
df_one_hot_encoding = pd.get_dummies(df_training[df_cat], 
                                    columns=["clarity","color","cut"],
                                    dtype=float)
df_one_hot_encoding
```

```python
diamonds_one_hot = pd.concat([df_num, df_one_hot_encoding], axis=1)
diamonds_one_hot.info()
```

```python
diamonds_one_hot['cut/weight_Fair']=diamonds_one_hot['cut_Fair']/diamonds_one_hot['carat']
diamonds_one_hot['cut/weight_Good']=diamonds_one_hot['cut_Good']/diamonds_one_hot['carat']
diamonds_one_hot['cut/weight_Ideal']=diamonds_one_hot['cut_Ideal']/diamonds_one_hot['carat']
diamonds_one_hot['cut/weight_Premium']=diamonds_one_hot['cut_Premium']/diamonds_one_hot['carat']
diamonds_one_hot['cut/weight_Very Good']=diamonds_one_hot['cut_Very Good']/diamonds_one_hot['carat']
diamonds_one_hot['color/weight_D']=diamonds_one_hot['color_D']/diamonds_one_hot['carat']
diamonds_one_hot['color/weight_E']=diamonds_one_hot['color_E']/diamonds_one_hot['carat']
diamonds_one_hot['color/weight_F']=diamonds_one_hot['color_F']/diamonds_one_hot['carat']
diamonds_one_hot['color/weight_G']=diamonds_one_hot['color_G']/diamonds_one_hot['carat']
diamonds_one_hot['color/weight_H']=diamonds_one_hot['color_H']/diamonds_one_hot['carat']
diamonds_one_hot['color/weight_I']=diamonds_one_hot['color_I']/diamonds_one_hot['carat']
diamonds_one_hot['color/weight_J']=diamonds_one_hot['color_J']/diamonds_one_hot['carat']
diamonds_one_hot['clarity/weight_I1']=diamonds_one_hot['clarity_I1']/diamonds_one_hot['carat']
diamonds_one_hot['clarity/weight_IF']=diamonds_one_hot['clarity_IF']/diamonds_one_hot['carat']
diamonds_one_hot['clarity/weight_SI1']=diamonds_one_hot['clarity_SI1']/diamonds_one_hot['carat']
diamonds_one_hot['clarity/weight_SI2']=diamonds_one_hot['clarity_SI2']/diamonds_one_hot['carat']
diamonds_one_hot['clarity/weight_VS1']=diamonds_one_hot['clarity_VS1']/diamonds_one_hot['carat']
diamonds_one_hot['clarity/weight_VS2']=diamonds_one_hot['clarity_VS2']/diamonds_one_hot['carat']
diamonds_one_hot['clarity/weight_VVS1']=diamonds_one_hot['clarity_VVS1']/diamonds_one_hot['carat']
diamonds_one_hot['clarity/weight_VVS2']=diamonds_one_hot['clarity_VVS2']/diamonds_one_hot['carat']
diamonds_one_hot
```

```python
diamonds_one_hot = diamonds_one_hot.drop(columns=['clarity_I1',
'clarity_IF',
'clarity_SI1',
'clarity_SI2',
'clarity_VS1',
'clarity_VS2',
'clarity_VVS1',
'clarity_VVS2',
'color_D',
'color_E',
'color_F',
'color_G',
'color_H',
'color_I',
'color_J',
'cut_Fair',
'cut_Good',
'cut_Ideal',
'cut_Premium',
'cut_Very Good'], axis=1)

diamonds_one_hot
```

```python
diamonds_one_hot['volume'] = diamonds_one_hot['x']*diamonds_one_hot['y']*diamonds_one_hot['z']
diamonds_one_hot
```
