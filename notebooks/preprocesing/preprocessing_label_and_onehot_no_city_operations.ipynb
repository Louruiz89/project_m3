```python
#importo las librerias necesarias

import pandas as pd
```

```python
#Trabajo sobre el dataset de training

df_final = pd.read_csv('./../data/diamons_clean.csv')
df_final.head()
```

```python
#reviso el dataframe que esté todo correcto

df_final.info()
```

```python
#divido el dataset en 2: numérico y categórico

df_num = df_final[["price","carat","depth","table","x","y","z"]]
df_cat =df_final[["clarity","color","cut"]]
```

```python
#Aplico One Hot enconding en las columnas categóricas

df_cat = ["clarity","color"]
df_one_hot_encoding = pd.get_dummies(df_final[df_cat], 
                                    columns=["clarity","color"],
                                    dtype=float)
df_one_hot_encoding
```

```python
diamonds_one_hot = pd.concat([df_num, df_one_hot_encoding,df_final['cut']], axis=1)
diamonds_one_hot.info()
```

```python
diamonds_one_hot['cut'].unique()
```

```python
cut_col  = {'Premium':5, 'Fair':4, 'Very Good':3, 'Good':2, 'Ideal':1}
def ordinal_encoding(x):
    for key in cut_col:
        if x == key:
            return cut_col[key]
```

```python
diamonds_one_hot['cut'] = diamonds_one_hot['cut'].apply(ordinal_encoding)
diamonds_one_hot
```

```python
diamonds_one_hot['color/weight_D']=diamonds_one_hot['color_D']/diamonds_one_hot['carat']
diamonds_one_hot['color/weight_E']=diamonds_one_hot['color_E']/diamonds_one_hot['carat']
diamonds_one_hot['color/weight_F']=diamonds_one_hot['color_F']/diamonds_one_hot['carat']
diamonds_one_hot['color/weight_G']=diamonds_one_hot['color_G']/diamonds_one_hot['carat']
diamonds_one_hot['color/weight_H']=diamonds_one_hot['color_H']/diamonds_one_hot['carat']
diamonds_one_hot['color/weight_I']=diamonds_one_hot['color_I']/diamonds_one_hot['carat']
diamonds_one_hot['color/weight_J']=diamonds_one_hot['color_J']/diamonds_one_hot['carat']
diamonds_one_hot['clarity/weight_I1']=diamonds_one_hot['clarity_I1']/diamonds_one_hot['carat']
diamonds_one_hot['clarity/weight_IF']=diamonds_one_hot['clarity_IF']/diamonds_one_hot['carat']
diamonds_one_hot['clarity/weight_SI1']=diamonds_one_hot['clarity_SI1']/diamonds_one_hot['carat']
diamonds_one_hot['clarity/weight_SI2']=diamonds_one_hot['clarity_SI2']/diamonds_one_hot['carat']
diamonds_one_hot['clarity/weight_VS1']=diamonds_one_hot['clarity_VS1']/diamonds_one_hot['carat']
diamonds_one_hot['clarity/weight_VS2']=diamonds_one_hot['clarity_VS2']/diamonds_one_hot['carat']
diamonds_one_hot['clarity/weight_VVS1']=diamonds_one_hot['clarity_VVS1']/diamonds_one_hot['carat']
diamonds_one_hot['clarity/weight_VVS2']=diamonds_one_hot['clarity_VVS2']/diamonds_one_hot['carat']
diamonds_one_hot
```

```python
diamonds_one_hot = diamonds_one_hot.drop(columns=['clarity_I1',
'clarity_IF',
'clarity_SI1',
'clarity_SI2',
'clarity_VS1',
'clarity_VS2',
'clarity_VVS1',
'clarity_VVS2',
'color_D',
'color_E',
'color_F',
'color_G',
'color_H',
'color_I',
'color_J'], axis=1)

diamonds_one_hot
```

```python
diamonds_one_hot['volume'] = diamonds_one_hot['x']*diamonds_one_hot['y']*diamonds_one_hot['z']
diamonds_one_hot
```
